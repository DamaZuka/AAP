### **1. Estrutura e Arquitetura do Sistema**

O sistema foi desenvolvido seguindo uma abordagem orientada a objetos, permitindo a separação clara entre a representação do estado do problema, a lógica de exploração do grafo e a experimentação algorítmica. A implementação divide-se em quatro componentes modulares fundamentais: `Node`, `PuzzleGraph`, `Algorithms` e o script de orquestração `main.py`.

###**2. Representação do Estado (Classe `Node`)**
A classe `Node` constitui o alicerce da representação do espaço de estados. Esta estrutura encapsula:
* **Estado do Tabuleiro:** Armazenado como uma tupla imutável, permitindo a sua utilização em estruturas de dados como conjuntos (`sets`) e dicionários, essenciais para a verificação de estados visitados.
* **Gestão de Caminho:** Cada nó mantém uma referência para o seu progenitor (`parent`), a ação executada para o alcançar e a sua profundidade na árvore de procura.
* **Métodos de Transição:** O método `get_successors` implementa a lógica de movimento do espaço vazio (valor 0), garantindo a validade das transições dentro dos limites do tabuleiro, calculados dinamicamente para suportar tanto o 8-Puzzle quanto o 15-Puzzle.
* **Heurísticas:** A classe integra os cálculos para as heurísticas de "peças fora do lugar" e "Distância de Manhattan", esta última computada através da diferença de coordenadas ortogonais entre a posição atual e a posição alvo de cada peça.


### **3. Definição do Espaço de Procura (Classe `PuzzleGraph`)**
A classe `PuzzleGraph` atua como a interface entre os algoritmos de procura e o domínio do problema. Esta abstração centraliza:
* **Gerador de Sucessores:** Delega a geração de estados vizinhos diretamente para a classe `Node`, encapsulando a lógica de transição.
* **Definição de Custo:** Define o custo de transição entre nós, que no caso do puzzle deslizante é constante e unitário (custo 1 por movimento), simplificando a lógica de expansão para algoritmos que consideram o custo do caminho.


### **4. Estratégias de Procura (Classe `Algorithms`)**
O módulo `Algorithms` implementa as diferentes estratégias de procura utilizadas na experimentação:
* **Procura Não Informada:** Inclui o *Depth-First Search* (DFS), implementado de forma recursiva, e o *Breadth-First Search* (BFS), utilizando uma fila (*queue*) para garantir a exploração por níveis e, consequentemente, a optimalidade da solução.
* **Procura Informada:** Implementa o *Greedy Best-First Search*, que seleciona o próximo nó com base exclusiva na função heurística, e o algoritmo *A\**, que combina o custo acumulado ($g(n)$) com a estimativa heurística ($h(n)$) para garantir a descoberta do caminho ótimo.
* **Controlo de Execução:** Todos os métodos integrados nestas classes incluem mecanismos de salvaguarda, como a monitorização do tempo de execução (*timeout*) e a limitação do número máximo de nós explorados, assegurando a estabilidade do sistema perante a explosão combinatória do espaço de estados.


### **5. Orquestração Experimental (`main.py`)**
O script `main.py` é responsável pela execução dos testes e pela persistência dos dados. Este componente realiza:
* **Configuração de Parâmetros:** Definição dos estados iniciais para o 8-Puzzle e 15-Puzzle, bem como a atribuição das heurísticas específicas para os testes informados.
* **Monitorização:** A função `run_experiment` mede o tempo de processamento e contabiliza os nós explorados, garantindo a uniformidade na recolha de métricas.
* **Persistência:** A funcionalidade `log_to_csv` automatiza a exportação dos resultados para um ficheiro estruturado (`resultados_parte_a.csv`), permitindo a análise posterior dos dados.


### **6. Metodologia de Análise e Visualização de Resultados**
Para a interpretação dos dados experimentais, foi desenvolvida uma interface interativa recorrendo às bibliotecas `pandas` e `ipywidgets`. Este componente processa o ficheiro `resultados_parte_a.csv`, realizando o tratamento de valores nulos e a normalização de tipos de dados para garantir a integridade da análise. Através de menus *dropdown*, é possível filtrar dinamicamente o desempenho por algoritmo, heurística ou nível de dificuldade, permitindo uma comparação granular entre as diferentes abordagens. Esta metodologia facilita a identificação de padrões de convergência e a deteção de estrangulamentos computacionais em estados de maior complexidade.



````python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# 1. Definição dos nomes das colunas
colunas = ["Algoritmo", "Heuristica", "Dificuldade", "Tempo(s)", "Estados_Explorados", "Comprimento_Solucao"]

# 2. Leitura do ficheiro
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AAP/resultados_parte_a.csv', names=colunas, header=0)

# 3. Limpeza de dados: Tratamento de valores nulos e conversão para string
# Isto é essencial para evitar erros de ordenação (sorting)
for col in ['Algoritmo', 'Heuristica', 'Dificuldade']:
    df[col] = df[col].fillna('N/A').astype(str)

# 4. Exibição da tabela inicial
display(df)

# 5. Configuração dos widgets de filtro
alg_select = widgets.Dropdown(options=['Todos'] + sorted(df['Algoritmo'].unique().tolist()), description='Algoritmo:')
heur_select = widgets.Dropdown(options=['Todos'] + sorted(df['Heuristica'].unique().tolist()), description='Heuristica:')
diff_select = widgets.Dropdown(options=['Todos'] + sorted(df['Dificuldade'].unique().tolist()), description='Dificuldade:')

# 6. Função de filtragem
def filtrar(algoritmo, heuristica, dificuldade):
    dados = df.copy()

    if algoritmo != 'Todos':
        dados = dados[dados['Algoritmo'] == algoritmo]
    if heuristica != 'Todos':
        dados = dados[dados['Heuristica'] == heuristica]
    if dificuldade != 'Todos':
        dados = dados[dados['Dificuldade'] == dificuldade]

    if dados.empty:
        print("Não foram encontrados resultados para esta combinação de filtros.")
    else:
        display(dados)

# 7. Apresentação da interface interativa
ui = widgets.VBox([alg_select, heur_select, diff_select])
out = widgets.interactive_output(filtrar, {'algoritmo': alg_select, 'heuristica': heur_select, 'dificuldade': diff_select})

display(ui, out)
````

Estes dados corroboram que, em espaços de estados massivos, a descoberta de soluções ótimas torna-se temporalmente inviável sem o suporte de recursos computacionais superiores ou heurísticas altamente refinadas que mitiguem este crescimento fatorial.

Ao capturar a geometria ortogonal do tabuleiro e a distância mínima que cada peça deve percorrer até à sua posição alvo, a métrica de Manhattan consegue 'dominar' a heurística de peças fora do lugar, reduzindo drasticamente o fator de ramificação efetivo e o volume de estados expandidos na fronteira. Esta limitação conduz frequentemente à necessidade de aceitar um **compromisso entre eficácia e optimalidade** (Pergunta 4). Como observado, o algoritmo *Greedy BFS* privilegia a rapidez de execução, conseguindo resolver estados complexos em frações de segundo, mas falha em garantir o caminho mais curto por ignorar o custo acumulado do trajeto. A **escalabilidade dos métodos** (Pergunta 5) foi testada na transição do 8-Puzzle para o 15-Puzzle, evidenciando que o aumento do número de peças provoca uma explosão combinatória massiva, elevando as configurações possíveis para a ordem dos $10^{13}$ estados. Esta transição torna a descoberta de soluções ótimas em estados muito desordenados um desafio temporalmente inviável para algoritmos sistemáticos sem o apoio de heurísticas compostas. Por fim, importa notar que a **garantia de uma solução ótima** (Pergunta 6) reside na mecânica dos algoritmos: o BFS garante a optimalidade em grafos de custo uniforme, enquanto o A* assegura o mesmo resultado desde que utilize uma heurística admissível, como as que foram implementadas e validadas neste trabalho.



##**PARTE B: Problema das N-Rainhas**


### **1. Implementação Técnica e Arquitetura do Código**

A resolução do problema das N-Rainhas foi estruturada em torno da classe `NQueensState`, que modela o tabuleiro de forma otimizada através de um array unidimensional de tamanho $N$. Nesta representação, o índice do array corresponde à coluna e o valor à linha ocupada pela rainha, eliminando nativamente conflitos na mesma coluna. O núcleo lógico da classe reside no método `get_conflicts`, que avalia a qualidade do estado calculando ataques em linhas e diagonais, e no método `get_neighbors`, que gera o espaço de vizinhança movendo rainhas verticalmente nas suas colunas.

A biblioteca de algoritmos implementada separa as abordagens em três categorias técnicas: as **determinísticas de vizinhança exaustiva** (*Hill Climbing*, *Tabu Search*), que avaliam todos os $N(N-1)$ vizinhos; as **paralelas/estocásticas** (*Stochastic Beam Search*), que mantêm um feixe de estados em simultâneo; e as de **amostragem** (*Stochastic Hill Climbing*), criadas para mitigar o custo computacional $O(N^4)$ em tabuleiros de dimensão $N=50$ ao avaliar apenas uma amostra aleatória da vizinhança. Toda a orquestração é gerida pela função `correr_experiencia` no script principal, que padroniza a recolha de métricas de sucesso, tempo e convergência iterativa, servindo de base para a análise experimental que se segue.




### **2. Metodologia Experimental e Limitações Computacionais**

Para a validação do desempenho dos algoritmos de procura local, estabeleceu-se um protocolo experimental que abrangeu três dimensões distintas de tabuleiro ($N=8$, $N=20$ e $N=50$). De forma a garantir a relevância estatística das métricas de sucesso e tempo, cada configuração foi submetida a um ciclo de 10 execuções independentes, partindo de estados iniciais gerados aleatoriamente.

A execução prática das experiências confirmou que a viabilidade da procura é condicionada pelo aumento drástico do espaço de estados e pelo custo de avaliação da vizinhança. Perante este cenário, a metodologia de testes foi adaptada no script de orquestração para aplicar restrições diferenciadas consoante a escala do problema:

* **Instâncias de Baixa e Média Escala ($N=8$ e $N=20$):** Os algoritmos operaram com limites de iterações permissivos (entre 500 a 1000), permitindo observar a convergência e a eficácia das heurísticas e dos parâmetros de controlo, como a temperatura e a lista tabu.
* **Instâncias de Larga Escala ($N=50$):** Face ao custo computacional proibitivo das transições exaustivas — que escalam numa ordem de $O(N^4)$ — a execução de algoritmos determinísticos revelou-se temporalmente inviável para o hardware disponível. Como estratégia de mitigação, optou-se pela implementação e utilização do Stochastic Hill Climbing. Em vez de avaliar a vizinhança completa de cada estado, este algoritmo recorre a uma amostragem estocástica (num_samples=100), permitindo explorar o espaço de estados de forma eficiente e realizar 200 iterações num tempo útil (aprox. 0.52s), algo que seria impossível com a abordagem clássica.

Nas instâncias de pequena e média complexidade ($N=8$ e $N=20$), observa-se uma superioridade inequívoca das abordagens que integram mecanismos de memória ou paralelismo heurístico. O Tabu Search e o Stochastic Beam Search demonstraram uma robustez notável, atingindo sistematicamente uma taxa de sucesso de 100%, mantendo a eficácia mesmo perante a expansão do espaço de estados. ### **5. Respostas às Perguntas do Enunciado**

A análise comparativa dos algoritmos de procura local aplicados ao problema das N-Rainhas demonstra que a dificuldade da resolução escala de forma combinatória e exponencial com o aumento de $N$. O espaço total de soluções válidas cresce drasticamente, mas o principal obstáculo reside na proliferação de ótimos locais profundos e na expansão quadrática da vizinhança. Como se verificou nos testes de $N=50$, os algoritmos que recorrem a um cálculo de vizinhança exaustivo, na ordem de $O(N^4)$, atingem rapidamente constrangimentos da arquitetura computacional, obrigando à sua substituição por heurísticas puramente estocásticas de amostragem.

É exatamente devido a este espaço de estados altamente acidentado que o *Hill Climbing* clássico falha de forma tão frequente. Este método rege-se puramente por uma exploração *greedy* de melhoria contínua e, por não possuir qualquer mecanismo algorítmico para aceitar temporariamente estados piores, acaba por ficar permanentemente retido no primeiro ótimo local que encontra, tornando-se incapaz de prosseguir a procura pelo ótimo global.

Para contornar esta estagnação, métodos como o *Simulated Annealing* adotam um mecanismo estocástico inspirado na termodinâmica, o qual permite aceitar transições para estados com pior valor heurístico. A probabilidade de aceitar um movimento negativo é elevada nas fases iniciais de alta temperatura, o que permite uma ampla exploração do tabuleiro, decaindo depois exponencialmente ao longo das iterações para forçar uma convergência fina, ou explotação, nas fases finais do algoritmo.

Outra forma de contornar os ótimos locais é garantindo a diversidade da procura, como acontece no *Stochastic Beam Search*, onde o parâmetro *beam width* dita o nível de paralelismo heurístico. Os resultados corroboram que um valor demasiado pequeno aproxima o comportamento do algoritmo ao do *Hill Climbing* clássico, aumentando o risco de confinamento, visto que uma largura de $W=2$ rendeu apenas 30% de sucesso. Em contraste, uma largura superior promove a manutenção da diversidade, preservando múltiplas linhas de evolução viáveis no espaço de estados; com $W=5$ a taxa de sucesso subiu para 100%, melhorando drasticamente as probabilidades de resolução em detrimento de uma maior utilização de memória.

De todos os métodos analisados, o *Tabu Search* destaca-se por melhorar consistentemente os resultados, demonstrando uma taxa de sucesso de 100% para as instâncias viáveis de $N=8$ e $N=20$. A inclusão de uma estrutura de memória, denominada lista tabu, previne que o algoritmo reverta imediatamente para configurações previamente visitadas, quebrando de forma muito eficaz as estruturas cíclicas que habitualmente afetam os métodos de procura local.

Consequentemente, é precisamente o *Tabu Search*, quando devidamente calibrado com um *Tabu Size* de 10, que apresenta o melhor compromisso entre tempo de execução e qualidade da solução. Para as instâncias de $N=20$, este algoritmo conseguiu resolver a totalidade das execuções de forma perfeita (0 conflitos) com um tempo médio de apenas 0.3759s e necessitando, em média, de 21.4 iterações. Revela-se, assim, largamente superior às restantes abordagens determinísticas, apresentando uma estabilidade de convergência muito mais robusta face a métodos puramente estocásticos como o *Simulated Annealing*.

## **PARTE C: Discussão Crítica e Reflexão Global**


####**1. Comparação de Algoritmos e Avaliação de Vantagens e Limitações**

A análise intra-problema revelou contrastes significativos no comportamento estrutural das diferentes estratégias de procura. No domínio do *Sliding Puzzle*, a procura não informada, nomeadamente o *Breadth-First Search* (BFS), garante a optimalidade da solução. Contudo, a sua principal limitação reside no crescimento exponencial da fronteira de nós mantidos em memória, tornando-o rapidamente inviável para instâncias mais profundas. Por sua vez, o *Greedy Search* apresentou uma vantagem inegável na rapidez de execução, pecando severamente pela sub-optimalidade das soluções. O algoritmo A* estabeleceu-se como o melhor compromisso sistemático, oferecendo soluções ótimas ao conciliar o custo acumulado com a estimativa heurística, não obstante partilhar a vulnerabilidade à exaustão de memória perante espaços de estados massivos.

Relativamente ao problema das N-Rainhas (procura local), o *Hill Climbing* evidenciou extrema eficiência computacional em tabuleiros pequenos, mas demonstrou como grave limitação a sua rápida e sistemática estagnação em ótimos locais. O algoritmo *Simulated Annealing* apresentou a vantagem teórica de mitigar esta estagnação ao permitir uma exploração termodinâmica flexível, exigindo, porém, uma calibração paramétrica minuciosa para garantir o sucesso. O *Tabu Search* revelou-se a abordagem mais robusta neste contexto, utilizando memória temporal para impedir configurações cíclicas e quebrar a convergência prematura. Finalmente, o *Stochastic Beam Search* providenciou a vantagem de um paralelismo heurístico eficiente, suportando a manutenção de diversidade através do sacrifício de uma maior alocação de memória transversal.

####**2. O Impacto das Heurísticas**
As heurísticas demonstraram ser o motor de tração fundamental da eficácia nos algoritmos informados. Na resolução do puzzle deslizante, a métrica da Distância de Manhattan superou largamente a heurística de "peças fora do lugar". Ao incorporar a verdadeira natureza espacial do problema (distância ortogonal das peças ao seu destino), guiou a exploração do A* de forma acentuadamente mais incisiva, resultando numa redução drástica dos estados expandidos na fronteira. Constata-se que a qualidade matemática e a proximidade da heurística ao custo real definem diretamente a exequibilidade computacional da procura, transformando problemas intratáveis numa navegação exequível.

####**3. Reflexão: Procura Clássica vs. Procura Local**
A transição entre abordagens sistemáticas e locais não é arbitrária, dependendo intrinsecamente da tipologia do problema e do propósito da solução. A procura clássica deverá ser empregue em domínios onde o caminho para atingir a meta constitui a própria solução (e.g., a sequência exata de movimentos num puzzle) e em contextos que exijam garantias absolutas de completude e optimalidade.

Por oposição, a procura local consolida-se como o paradigma indicado quando apenas o estado final é relevante para a resolução (e.g., uma disposição válida de rainhas no tabuleiro), não importando o histórico de estados alcançados. Esta metodologia abdica da completude em prol de uma gestão de memória minimalista, frequentemente $O(1)$, tornando-se a ferramenta de eleição para problemas de otimização em espaços combinatórios massivos onde a sistematização é inexequível.


####**4. Escalabilidade dos Métodos**
A capacidade de escalar face ao aumento de complexidade delimitou claramente o alcance prático destas metodologias. Os métodos de procura clássica evidenciaram limitações severas de escalabilidade espacial; a transição entre o 8-Puzzle e o 15-Puzzle gerou uma explosão combinatória que rapidamente saturou a capacidade de processamento iterativo.

Em contraste, os algoritmos de procura local escalam de forma notável no vetor da memória, mas demonstraram ser suscetíveis a restrições temporais caso a complexidade da função de vizinhança seja demasiado elevada. A experimentação evidenciou que instâncias como $N=50$ inflacionam o custo de avaliação para a ordem de $O(N^4)$, precipitando estrangulamentos críticos que exigem forçosamente a introdução de amostragem estocástica (como o *Stochastic Hill Climbing*) para restabelecer a escalabilidade temporal da exploração.

#**PROBLEMA 2: SIMPLEX**

Nesta secção, aborda-se a resolução do Problema 2, focado na otimização do planeamento de produção diária de refeições de um restaurante. O objetivo central consiste em maximizar o lucro diário através da determinação da quantidade ideal a confecionar de três pratos distintos: frango grelhado, massa vegetariana e salmão no forno. Esta maximização está sujeita a limitações estritas de recursos operacionais, nomeadamente o tempo de preparação, a disponibilidade de ingredientes frescos e a capacidade de empratamento. Para a resolução analítica deste problema de programação linear, aplicar-se-á o método Simplex.

###**1.Formulação do Problema**
**Variáveis de Decisão:**
* **$x_1$:** número de pratos de frango grelhado
* $x_2$: número de pratos de massa vegetariana
* $x_3$: número de pratos de salmão no forno

**Função Objetivo:**

Pretende-se maximizar o lucro diário ($Z$):$$Z = 6x_1 + 7x_2 + 6x_3$$
**Restrições Operacionais:**

As limitações diárias de recursos traduzem-se nas seguintes inequações:
* $$1x_1 + 2x_2 + 1x_3 \le 60$$(Restrição de tempo de preparação
* $$2x_1 + 1x_2 + 3x_3 \le 95$$(Restrição de ingredientes frescos)
* $$2x_1 + 2x_2 + 1x_3 \le 70$$(Restrição de capacidade de empratamento)
* $$x_1, x_2, x_3 \ge 0$$(Condição de não-negatividade)

Nesta etapa inicial, procedeu-se à abstração matemática do problema real enunciado. As variáveis de decisão ($x_1$, $x_2$ e $x_3$) foram estipuladas para representar as incógnitas do problema, ou seja, as quantidades exatas de cada prato a produzir. A função objetivo ($Z$) foi estruturada associando o lucro unitário de cada prato (6€, 7€ e 6€, respetivamente) à sua variável de decisão correspondente, estabelecendo a meta de maximização. Por fim, as restrições foram equacionadas para garantir que o consumo de recursos inerente à produção planeada não excede os limites máximos diários estabelecidos (60 unidades de tempo, 95 unidades de ingredientes e 70 unidades de capacidade de empratamento). A restrição de não-negatividade foi incluída por imperativo lógico, dado que é impossível confecionar quantidades negativas de refeições.

Após a definição do modelo matemático, o passo seguinte para a aplicação do algoritmo Simplex consiste na conversão do problema para a sua forma standard. Este processo é essencial para transformar as inequações das restrições em equações lineares, permitindo a construção do quadro Simplex (tableau) e a execução das operações matriciais necessárias para encontrar a solução ótima

###**2.Passagem à Forma StandardIntrodução de Variáveis de Folga:**
Para converter as desigualdades do tipo $\le$ em igualdades, introduzem-se variáveis de folga ($s_1, s_2, s_3$), que representam a quantidade de recurso não utilizada em cada restrição.  Função Objetivo em Forma de Equação:
A função objetivo é reescrita de modo a que o lado direito da equação seja igual a zero, mantendo o coeficiente da variável $Z$ positivo:
  $$Z - 6x_1 - 7x_2 - 6x_3 = 0$$Sistema de Equações (Forma Standard):
O modelo completo na forma standard é apresentado da seguinte forma:
Maximizar $Z$, sujeito a:  $$x_1 + 2x_2 + x_3 + s_1 = 60$$$$2x_1 + x_2 + 3x_3 + s_2 = 95$$$$2x_1 + 2x_2 + x_3 + s_3 = 70$$$$x_1, x_2, x_3, s_1, s_2, s_3 \ge 0$$

Nesta fase, o modelo foi preparado para o processamento algorítmico através de duas modificações principais. Primeiro, a função objetivo foi reestruturada para integrar uma equação de balanço nulo, o que facilita a identificação dos coeficientes de lucro no quadro Simplex. Segundo, foram adicionadas três variáveis de folga, $s_1, s_2$ e $s_3$, correspondentes às restrições de tempo de preparação, ingredientes frescos e capacidade de empratamento, respetivamente.  Estas variáveis de folga assumem um papel crítico na interpretação física da solução, pois indicam o volume de recursos remanescentes em cada etapa da produção. Com todas as restrições convertidas em igualdades e garantindo que todas as variáveis (decisão e folga) são não-negativas, o problema atinge a conformidade com os requisitos formais do método Simplex.

Com o modelo matemático estabelecido na sua forma standard, procede-se à aplicação sistemática do algoritmo Simplex. Este método iterativo explora os vértices da região viável de soluções, movendo-se progressivamente de uma solução básica admissível para outra que ofereça um valor superior para a função objetivo, até que seja impossível obter melhorias adicionais. Abaixo detalha-se a primeira iteração algorítmica, seguida de um sumário das iterações subsequentes que conduzem à otimização final.

###**3.Resolução do problema pelo método Simplex**
**Construção do Quadro Simplex Inicial:**


| Base | $x_1$ | $x_2$ | $x_3$ | $s_1$ | $s_2$ | $s_3$ | Solução |
| :--- | :--- | :--- | :--- | :--- | :--- | :--- | :--- |
| **$s_1$** | 1 | 2 | 1 | 1 | 0 | 0 | 60 |
| **$s_2$** | 2 | 1 | 3 | 0 | 1 | 0 | 95 |
| **$s_3$** | 2 | 2 | 1 | 0 | 0 | 1 | 70 |
| **$Z$** | -6 | -7 | -6 | 0 | 0 | 0 | 0 |

**Primeira Iteração:**

 1. **Identificação da Coluna Pivô (Variável a entrar na base):**
  A variável que entra na base é $x_2$, por possuir o coeficiente mais negativo na linha da função objetivo ($Z$). O valor é -7.

 2. **Identificação da Linha Pivô (Variável a sair da base):** Calculam-se os rácios entre a coluna das constantes (Solução) e os elementos estritamente positivos da coluna pivô ($x_2$):
    * Linha $s_1$: $\frac{60}{2} = 30$
    * Linha $s_2$: $\frac{95}{1} = 95$
    * Linha $s_3$: $\frac{70}{2} = 35$
   
    O menor rácio estritamente positivo é 30, originado pela linha $s_1$. Deste modo, a variável $s_1$ sai da base. O elemento pivô corresponde ao valor 2 (interseção da coluna $x_2$ com a linha $s_1$).

  3. **Operações sobre as Linhas (Atualização do Quadro):**
      * Nova linha $x_2$: Obtida pela divisão da antiga linha $s_1$ pelo pivô (2).$$L_1' = \frac{L_1}{2} = \left[ \frac{1}{2}, 1, \frac{1}{2}, \frac{1}{2}, 0, 0 \mid 30 \right]$$

      * Nova linha $s_2$: Antiga linha $s_2 - (1 \times L_1')$.$$L_2' = L_2 - 1L_1' = \left[ \frac{3}{2}, 0, \frac{5}{2}, -\frac{1}{2}, 1, 0 \mid 65 \right]$$

      * Nova linha $s_3$: Antiga linha $s_3 - (2 \times L_1')$.$$L_3' = L_3 - 2L_1' = \left[ 1, 0, 0, -1, 0, 1 \mid 10 \right]$$

      * Nova linha $Z$: Antiga linha $Z - (-7 \times L_1')$.$$L_Z' = L_Z + 7L_1' = \left[ -\frac{5}{2}, 0, -\frac{5}{2}, \frac{7}{2}, 0, 0 \mid 210 \right]$$
      



 **Quadro Simplex após a 1ª Iteração:**
 | Base | $x_1$ | $x_2$ | $x_3$ | $s_1$ | $s_2$ | $s_3$ | Solução |
| :--- | :--- | :--- | :--- | :--- | :--- | :--- | :--- |
| **$x_2$** | 1/2 | 1 | 1/2 | 1/2 | 0 | 0 | 30 |
| **$s_2$** | 3/2 | 0 | 5/2 | -1/2 | 1 | 0 | 65 |
| **$s_3$** | 1 | 0 | 0 | -1 | 0 | 1 | 10 |
| **$Z$** | -5/2 | 0 | -5/2 | 7/2 | 0 | 0 | 210 |

**Iterações Seguintes:**

* **Segunda Iteração:** Constata-se a presença de coeficientes negativos na linha $Z$ ($-\frac{5}{2}$ para $x_1$ e $x_3$). Optando pela entrada da variável $x_1$, a variável que sai da base é $s_3$ (devido ao menor rácio positivo, $\frac{10}{1} = 10$). Após o pivoteamento, o lucro na função objetivo ascende a 235€.
* **Terceira Iteração:** O último coeficiente negativo na linha $Z$ pertence à variável $x_3$, que entra na base. O cálculo dos rácios determina a saída da variável $s_2$. A execução destas operações matriciais elimina todos os valores negativos na linha $Z$, indicando que a solução ótima foi alcançada com um lucro de 285€.

### **4.Discussão Crítica e Interpretação da Solução**

**Análise da Execução do Algoritmo Simplex:**

Nesta etapa fundamental, o método Simplex foi aplicado manipulando o quadro (matriz) que representa o sistema de equações do problema. A iteração inicial revelou que a variável com maior impacto negativo no lucro não concretizado era $x_2$ (massa vegetariana), definindo a sua entrada na solução básica. A variável limitante para esta entrada era o tempo de preparação ($s_1$), que foi inteiramente consumido, saindo assim da base.

Através de operações elementares de linha baseadas em pivoteamento matricial ou eliminação de Gauss-Jordan, o quadro foi normalizado, resultando num aumento imediato do lucro para 210€.
Contudo, a presença continuada de coeficientes negativos na linha da função objetivo ($Z$) sinalizou que a configuração ainda não era a ideal. Deste modo, realizaram-se duas iterações subsequentes, incorporando sucessivamente as variáveis $x_1$ e $x_3$ na base produtiva. O algoritmo encerrou ao confirmar que não existiam mais valores negativos na linha de avaliação (linha $Z$), o que garante matematicamente a identificação do ponto de lucro máximo absoluto.

**Interpretação dos Resultados Otimizados:**

Com base no quadro final obtido, a solução ótima para a maximização do lucro diário do restaurante é interpretada da seguinte forma:

* **Plano de Produção:** Para maximizar o lucro, o restaurante deve preparar:
    * $10$ pratos de frango grelhado ($x_1 = 10$)
    * $15$ pratos de massa vegetariana ($x_2 = 15$)
    * $20$ pratos de salmão no forno ($x_3 = 20$)
* **Lucro Máximo:** Este planeamento resulta num lucro diário total de $285€$.
* **Utilização de Recursos:** Verifica-se que todas as variáveis de folga finais são nulas ($s_1 = 0, s_2 = 0, s_3 = 0$), indicando que o restaurante utiliza a totalidade das $60$ unidades de tempo, $95$ unidades de ingredientes e $70$ unidades de capacidade de empratamento disponíveis.


### **Validação Computacional com PuLP**

````python
#!pip install pulp
````

A biblioteca PuLP é um modelador de programação linear escrito em Python, amplamente utilizado tanto em ambiente académico como industrial para resolver problemas de otimização matemática. Esta ferramenta permite a definição intuitiva de variáveis de decisão, funções objetivo e restrições lineares, servindo como uma interface para diversos solvers (motores de cálculo) de alto desempenho. No contexto deste projeto, a utilização do PuLP é fundamental para a validação computacional dos resultados obtidos através do método Simplex manual, garantindo a precisão dos valores de produção encontrados e assegurando que a solução proposta corresponde, efetivamente, ao ótimo global do sistema

````python
from pulp import LpMaximize, LpProblem, LpVariable

# 1. Inicialização do modelo matemático
modelo = LpProblem("Maximizacao_Lucro_Restaurante", LpMaximize)

# 2. Definição das variáveis de decisão (assumindo a condição de não-negatividade)
x1 = LpVariable("Frango_Grelhado", lowBound=0)
x2 = LpVariable("Massa_Vegetariana", lowBound=0)
x3 = LpVariable("Salmao_No_Forno", lowBound=0)

# 3. Definição da Função Objetivo
modelo += 6 * x1 + 7 * x2 + 6 * x3, "Lucro_Diario"

# 4. Definição das Restrições Operacionais
modelo += 1 * x1 + 2 * x2 + 1 * x3 <= 60, "Tempo_de_Preparacao"
modelo += 2 * x1 + 1 * x2 + 3 * x3 <= 95, "Ingredientes_Frescos"
modelo += 2 * x1 + 2 * x2 + 1 * x3 <= 70, "Capacidade_de_Empratamento"

# 5. Execução do algoritmo de otimização
modelo.solve()

# 6. Apresentação formal dos resultados
print("Estado da Otimização:", modelo.status)
print(f"Lucro Máximo Atingido: {modelo.objective.value()} €")
print("\nPlano de Produção Diário:")
for variavel in modelo.variables():
    print(f"- {variavel.name}: {variavel.value()} pratos")
````

### **Conclusão e Validação dos Resultados**

Após a implementação computacional do modelo recorrendo à biblioteca `PuLP`, é possível confirmar a total convergência entre a resolução analítica e o resultado algorítmico. A execução do código validou as seguintes métricas:

* **Plano de Produção Ótimo:** Confirmou-se a necessidade de produzir $10$ pratos de frango grelhado, $15$ de massa vegetariana e $20$ de salmão no forno para atingir o desempenho máximo.
* **Lucro Máximo:** O valor de $285,0€$ apurado computacionalmente coincide com o resultado do método Simplex manual, validando a integridade de todos os cálculos de pivoteamento realizados.
* **Eficiência de Recursos:** O estado da otimização indicou que a solução encontrada é única e estável, com a saturação total dos recursos disponíveis ($s_1 = s_2 = s_3 = 0$).

Em suma, a dupla abordagem — manual e computacional — permite concluir com absoluta segurança que a solução apresentada é a mais eficiente dentro do espaço de estados definido pelas restrições do restaurante.
